### Umbralización para extraer las características 

En este markdown se analiza el proceso de umbralización con el objetivo de preparar la imagen en las mejores condiciones posibles para extraer sus características de forma adecuada y obtener resultados óptimos.

Al aplicar una umbralización adecuada, se evita la pérdida de detalles en las huellas, permitiendo una mejor definición de las crestas y valles. Por ello, se busca determinar el método de umbralización que proporcione el equilibrio más eficaz entre claridad y precisión.

```python
import cv2 as cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('../BaseDatosNISTsd09_pr/crd_0813f/crd_0813f_01.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

thr = cv2.adaptiveThreshold(gray, 200, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 10) 
# esta función realiza la umbralización adaptativa
# dicha función fue recomendada por ChatGPT tras analizar diferentes métodos de umbralización

plt.imshow(cv2.cvtColor(thr, cv2.COLOR_BGR2RGB))
plt.title('Puntos')
plt.show()
```
