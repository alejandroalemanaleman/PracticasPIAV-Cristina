```python
import os
import cv2 as cv
import numpy as np

def recortar_huella_adaptativo(img):
    """Recorta automáticamente la huella, ignorando ruido alrededor."""

    # Convertir a gris si es necesario
    if len(img.shape) == 3:
        img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

    # Umbralización (Otsu)
    _, thresh = cv.threshold(img, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)
    
    # Invertir si el fondo es blanco
    if np.mean(thresh[:10, :10]) > 127:
        thresh = cv.bitwise_not(thresh)

    # Sumar por filas y columnas
    sum_rows = np.sum(thresh, axis=1)
    sum_cols = np.sum(thresh, axis=0)

    # Calcular la media (solo donde hay píxeles > 0)
    mean_row = np.mean(sum_rows[sum_rows > 0])
    mean_col = np.mean(sum_cols[sum_cols > 0])

    # Umbral relativo (ajustable)
    thr_row = 0.1 * mean_row
    thr_col = 0.1 * mean_col

    # Detectar zonas donde hay suficiente densidad de píxeles
    rows_nonzero = np.where(sum_rows > thr_row)[0]
    cols_nonzero = np.where(sum_cols > thr_col)[0]

    if len(rows_nonzero) == 0 or len(cols_nonzero) == 0:
        return img  # no se detectó huella útil

    # Limites con pequeño margen
    margen = 10
    y1 = max(rows_nonzero[0] - margen, 0)
    y2 = min(rows_nonzero[-1] + margen, img.shape[0])
    x1 = max(cols_nonzero[0] - margen, 0)
    x2 = min(cols_nonzero[-1] + margen, img.shape[1])

    cropped = img[y1:y2, x1:x2]
    return cropped


# ==== DEMO CON VISUALIZACIÓN ====
sample_path = "BaseDatosNISTsd09_pr"
folders = sorted(os.listdir(sample_path))

for folder in folders:
    dir_path = os.path.join(sample_path, folder)
    files = sorted(os.listdir(dir_path))

    if len(files) == 0:
        continue

    ruta = os.path.join(dir_path, files[0])
    img = cv.imread(ruta, cv.IMREAD_GRAYSCALE)
    if img is None:
        continue

    cropped = recortar_huella_adaptativo(img)

    cv.imshow("Original", img)
    cv.imshow("Recortada", cropped)

    key = cv.waitKey(1500)
    if key == 27:  # ESC para salir
        break

cv.destroyAllWindows()
```
